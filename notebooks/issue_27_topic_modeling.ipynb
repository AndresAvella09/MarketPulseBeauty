> Este notebook documenta el modelo baseline de tópicos, con visualizaciones e interpretación preliminar.

```python
import pandas as pd
import matplotlib.pyplot as plt

topics = pd.read_csv("../data/processed/topics_summary.csv")
df = pd.read_csv("../data/processed/reviews_with_topics.csv")

print("topics:", topics.shape)
print("df:", df.shape)
display(topics)
display(df.head(3))
```

## 1) Distribución de tópicos
Mide cuántas reseñas quedaron asignadas a cada `topic_id` (tópicos dominantes).

```python
counts = df["topic_id"].value_counts().sort_index()
display(counts.to_frame("n_reviews"))

plt.figure()
counts.plot(kind="bar")
plt.title("Distribución de reseñas por topic_id")
plt.xlabel("topic_id")
plt.ylabel("Cantidad de reseñas")
plt.show()
```

## 3) Ejemplos de reseñas por tópico (muestra)
Se muestran ejemplos para validar que el `topic_id` sea coherente con los términos del tópico.

```python
cols = [c for c in ["topic_id","pd_id","keyword","Rating","sentiment_score","ReviewText"] if c in df.columns]

for t in sorted(df["topic_id"].unique()):
    print(f"\n=== TOPIC {t} ===")
    display(df[df["topic_id"] == t][cols].head(3))
```

## 4) Hallazgos preliminares
- Se identificaron **6 tópicos**. Los términos sugieren temas recurrentes como:
  - hidratación/piel seca (`skin`, `dry`, `cream`)
  - cuero cabelludo/serum (`scalp`, `serum`)
  - barrera/sensibilidad (`barrier`, `sensitive`)
  - rendimiento/cantidad (`goes long`, `little`)
  - cabello graso / lavados (`hair`, `oily`, `wash`)
  - skincare/acné (`moisturizer`, `acne`, `face`)
- La distribución de `topic_id` muestra qué tópicos concentran más reseñas.
- Próximo paso recomendado: cruzar `topic_id` con `Rating` o `sentiment_score` para ver qué tópicos se asocian a reseñas negativas/positivas.

```python
# Promedio rating por tópico 
if "Rating" in df.columns:
    by_topic_rating = df.groupby("topic_id")["Rating"].mean().sort_index()
    display(by_topic_rating.to_frame("avg_rating"))

    plt.figure()
    by_topic_rating.plot(kind="bar")
    plt.title("Promedio de Rating por topic_id")
    plt.xlabel("topic_id")
    plt.ylabel("Rating promedio")
    plt.show()

# Promedio sentiment por tópico 
if "sentiment_score" in df.columns:
    by_topic_sent = df.groupby("topic_id")["sentiment_score"].mean().sort_index()
    display(by_topic_sent.to_frame("avg_sentiment"))

    plt.figure()
    by_topic_sent.plot(kind="bar")
    plt.title("Promedio de sentiment_score por topic_id")
    plt.xlabel("topic_id")
    plt.ylabel("Sentiment promedio")
    plt.show()
```
